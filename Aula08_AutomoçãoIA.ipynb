```python
'''
Crie um script simples em python para verificar
a temperatura de um sistema. Se for maior que 30
exiba um alerta de superaquecimento.
'''

# Solicita ao usuário que insira a temperatura atual do sistema.
# O valor inserido é convertido para um número inteiro.
temperatura_atual = int(input("Digite a temperatura atual do sistema: "))

# Imprime a temperatura atual do sistema usando concatenação de strings
print("Temperatura atual do sistema: " + str(temperatura_atual) + "°C")

# Verifica se a temperatura atual é maior que 30
if temperatura_atual > 30:
    # Se for maior, exibe um alerta de superaquecimento
    print("ALERTA: Superaquecimento detectado!")
else:
    # Caso contrário, informa que a temperatura está normal
    print("Temperatura normal.")
```

```python
import os
import shutil
import random

# --- CONFIGURAÇÕES ---
PASTA_ORIGEM = 'arquivos_baguncados'
PASTA_DESTINO = 'arquivos_organizados'
EXTENSOES_TESTE = ['.pdf', '.jpg', '.png', '.doc', '.docx', '.txt', '.xlsx', '.mp3', '.mp4']

def preparar_ambiente():
    """Garante que as pastas base existam antes de começar."""
    for pasta in [PASTA_ORIGEM, PASTA_DESTINO]:
        if not os.path.exists(pasta):
            os.makedirs(pasta)
            print(f"✔ Diretório '{pasta}' preparado.")

def criar_arquivos_ficticios(quantidade=15):
    """Gera arquivos aleatórios para simular uma pasta bagunçada."""
    print(f"\n--- Gerando {quantidade} arquivos de teste ---")
    for i in range(1, quantidade + 1):
        extensao = random.choice(EXTENSOES_TESTE)
        nome_arquivo = f"arquivo_teste_{i}{extensao}"
        caminho_completo = os.path.join(PASTA_ORIGEM, nome_arquivo)

        with open(caminho_completo, 'w') as f:
            f.write(f"Conteúdo do arquivo {nome_arquivo}")
        print(f"📄 Criado: {nome_arquivo}")

def organizar_arquivos():
    """Lê a pasta de origem e move cada arquivo para uma subpasta baseada na extensão."""
    print("\n--- Iniciando Organização ---")

    # Filtra apenas arquivos, ignorando pastas que possam estar lá dentro
    arquivos = [f for f in os.listdir(PASTA_ORIGEM)
                if os.path.isfile(os.path.join(PASTA_ORIGEM, f))]

    if not arquivos:
        print("⚠ Nenhnum arquivo encontrado para organizar.")
        return

    for nome_arquivo in arquivos:
        caminho_origem = os.path.join(PASTA_ORIGEM, nome_arquivo)

        # Extrai a extensão e define o nome da subpasta (ex: '.pdf' -> 'pdf')
        _, extensao = os.path.splitext(nome_arquivo)
        nome_subpasta = extensao[1:].lower() if extensao else 'outros'

        caminho_subpasta = os.path.join(PASTA_DESTINO, nome_subpasta)

        # Cria a subpasta específica da extensão se não existir
        if not os.path.exists(caminho_subpasta):
            os.makedirs(caminho_subpasta)

        # Move o arquivo
        shutil.move(caminho_origem, os.path.join(caminho_subpasta, nome_arquivo))
        print(f"🚚 {nome_arquivo} -> /{nome_subpasta}/")

    print("\n✅ Organização concluída com sucesso!")

# --- EXECUÇÃO DO SCRIPT ---
if __name__ == "__main__":
    preparar_ambiente()
    criar_arquivos_ficticios(15)
    organizar_arquivos()
```
