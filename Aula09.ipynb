***PRÁTICAS, FERRAMENTAS DE IA E ÉTICA NO DESENVOLVIMENTO (UNIDADE II)***

**INTRODUÇÃO:**
Este parece ter como objetivo apresentar reflexões sobre a Unidade II, baseando-se no conteúdo abordado no Podcast da disciplina. O foco principal é a integração de assistentes de inteligência artificial no desenvolvimento de software e seus impactos nas práticas diárias e no fluxo de trabalho. É fundamental destacar as indicações de que a IA não é um oráculo e não é infalível. Conforme planejado, a IA não está aqui para substituir o programador, mas muito pelo contrário, busca-se estabelecer um cenário onde o programador e a IA se somam em vez de um atrapalhar o outro.

**AUTOMAÇÃO:** É evidente que a IA é excelente para criar código padrão (códigos repetitivos), permitindo focar na lógica principal. Como mencionado no podcast, você descreve um comentário e a ferramenta pode gerar todo esse esqueleto de arquivo e código. Em outras palavras, a IA te entrega o palco montado e o seu trabalho é só colocar os atores em cena.

**REFATORAÇÃO:** O uso dessas ferramentas para analisar um código existente e sugerir melhorias de desempenho ou legibilidade é uma das suas maiores vantagens.

**SUGESOTES:** Essas ferramentas funcionam como um complemento avançado, analisando o contexto do arquivo em tempo real. Na maior parte das vezes, a IA acerta bastante, sendo excelente para antecipar funções comuns e padrões repetitivos. No entanto, ela atrapalha em certas situações ao tentar adivinhar lógicas ou regras de negócio muito específicas do projeto, pode sugerir códigos incorretos que parecem certos num primeiro momento. Isso implica o programador a pausar seu raciocínio para depurar e corrigir o erro gerado, quebrando seu fluxo de pensamento.

**FLUXO E ÉTICA:**
A introdução dessas ferramentas não muda apenas o código, mas o fluxo de trabalho (fluxo de trabalho) do desenvolvedor como um todo. A transição de “escrever cada linha” para “revisar e aprovar o código gerado pela IA” exige uma nova habilidade de leitura crítica. Em relação à ética, o podcast clamou pontos cruciais. Ao decorrer disto: Não há ambiente ético e de segurança, é fundamental compreender que a responsabilidade final pelo código é desenvolvida pelo desenvolvedor. A IA atua como uma ferramenta de auxílio, não como um agente autônomo.

**CONCLUSÃO:**
De todo modo, as ferramentas de IA generativa para código são aliadas poderosas que vieram para ficar. Contudo, elas desabilitam que o profissional atue mais como um "arquiteto e revisor" do que como um mero digitador de código, mantendo sempre a responsabilidade ética e técnica sobre o produto final.
